```python
%matplotlib notebook
import numpy as np
import matplotlib.pyplot as plt
from transforms import random_poly, my_poly_features
```

```python
#synthetic data for demonstration purpose
x=np.linspace(-10,10,1000)
y=random_poly(x,5)
```

```python
#standardizing
x_mean=np.mean(x,axis=0,keepdims=True)
x_std=np.std(x,ddof=1,axis=0,keepdims=True)
x_standard=((x-x_mean)/x_std)[:,np.newaxis]
y_mean=np.mean(y)
y_std=np.std(y,ddof=1)
y_standard=((y-y_mean)/y_std)[:,np.newaxis]
```

```python
#plotting
plt.figure(1)
plt.plot(x_standard,y_standard,'rx')
plt.xlabel("Standardized x-values")
plt.ylabel("Standardized y-values")
```

```python
x_poly=my_poly_features(x_standard, 5)
x_poly.shape
params=np.linalg.inv(x_poly.T @ x_poly) @ x_poly.T @ y_standard
y_pred=x_poly @ params
```

```python
plt.figure(2)
plt.plot(x_standard[:,0],y_standard,'rx',label="Actual Data")
plt.plot(x_standard[:,0],y_pred,'b',label="Fitted Data")
plt.xlabel("Standardized x-values")
plt.ylabel("Standardized y-values")
plt.legend()
```
